## Linear Regression

is used to predict continuous values based on a line which best models the relationship between the independent X and the dependent variable Y which best fits a series of points. In other words, linear regression helps predict values on a continuous spectrum rather than discrete classes.

Linear regression is the simplest form of regression. The goal is to find the line parameterized by some set of slope and y-intercept parameters which provide the best fit to our training data.

How do we train a neural network to perform a linear regression?
It does by minimizing the loss over our fata sets which would correspond to the mean squared error.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
np.random.seed(0)
points = 500
X = np.linspace(-3, 3, points)
y = np.sin(X) + np.random.uniform(-0.5, 0.5, points)

plt.scatter(X,y)
```

### Why is our input dimension only 1? 

So the prediction (the output) is the quantitative response variable y which is dependent on the input, its corresponding explanatory variable x, the independent variable x. So the response variable is the dependent variable whose value we want to explain and predict based on the values of the independent variable x.

Accordingly, the only input in this regression type example, this very simple regression type example is going to be the x-value

### Mean Squared Error:

When we first started some regression line by parameterizing a random set of weights, to measure how well this model fits, we calculate the standard error of the estimate; The mean squared error.

![Mean Squared Error](https://i.imgur.com/kTIYRBX.jpg)

The standard error of the estimate is a measure of the accuracy of the models prediction.
The data points show the actual values, the actual labels of a given input, whereas the regression line shows the estimated value. For each actual value, there is a corresponding estimated value predicted by the linear model.

The error in this case corresponds to **how far the actual values is from the estimated value, which is just the vertical distance between them.** so the error is pretty much comparing the estimated values from the actual values.

The MSE can be define as the following formula:
Mean Squared Error Formula

Where we take the actual value and subtract it by the estimated value returned by the model; Take the square and repeating the process for every points, summing up all the values and then divided by the number of data points to obtain the total error.

```python
model = Sequential()
model.add(Dense(50, input_dim = 1, activation = 'sigmoid'))
model.add(Dense(30, activation = 'sigmoid'))
model.add(Dense(1))

adam = Adam(lr = 0.01)
model.compile(loss = 'mse')
```
